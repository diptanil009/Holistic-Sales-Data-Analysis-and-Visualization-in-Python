2. Sales Trend Analysis
Objective: Track sales performance over time to understand seasonal trends or growth.

```python
#convert 'Order Date' data type as datetime
data['Order Date'] = pd.to_datetime(data['Order Date'],format='mixed')
```

```python
data.dtypes
```

```python
# Group by month and sum the sales
monthly_sales= data.groupby(data['Order Date'].dt.to_period('M'))['Sales'].sum()
```

```python
monthly_sales
```

```python
# Plotting sales trends
monthly_sales.plot(title="Sales Trend Over Time", ylabel="Sales")
plt.show()
```

3. Top Customers Analysis
Objective: Identify the most valuable customers based on total sales.


```python
top_customers= data.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
top_customers
```

```python
top_customers.plot(kind='bar',title="Top 10 Customers by Sales")
plt.show()
```

4. Product Performance Analysis
   Objective: Identify the best and worst-performing products in terms of sales and profit.

```python
product_performance =data.groupby('Product Name',as_index=False)[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False).head(10)
```

```python
product_performance
```

```python
product_performance.plot(kind='bar',title='Top 10 Products by sales')
plt.show()
```

5. Discount vs Profit Analysis
   Objective: Investigate how different levels of discount affect profitability.

```python
import seaborn as sns
```

```python
# Plot relationship between discount and profit
sns.scatterplot(x='Discount',y='Profit',data=data)
plt.show()
```

6. Shipping Mode Efficiency
   Objective: Analyze the efficiency of different shipping modes based on cost and delivery times.

```python
data.columns
```

```python
shipping_mode=data.groupby('Ship Mode',as_index=False)[['Shipping Cost','Profit']].mean()
```

```python
shipping_mode
```

```python
shipping_mode.plot(kind='bar',title='Shipping Mode Efficiency')
plt.show()
```

7. Profit Margin Calculation
Objective: Calculate and analyze profit margins across various categories or segments.

```python
data['Profit Margin'] = (data['Profit']/ data['Sales'])*100
```

```python
profit_margin_by_category=data.groupby('Category')['Profit Margin'].mean()
```

```python
profit_margin_by_category
```

```python
profit_margin_by_category.plot(kind='bar',title= 'Average Profit Margin by Category')
plt.show()
```

8. Order Priority Analysis
 Objective: Analyze the relationship between Order Priority and sales performance.

```python
order_priority_sales=data.groupby('Order Priority')['Sales'].sum()
```

```python
order_priority_sales.plot(kind='pie',title='sales by Order Priority',autopct='%1.1f%%')
plt.show()
```

9. Segment-Wise Performance
    Objective: Examine sales and profit performance across different customer segments.

```python
segment_performence=data.groupby('Segment')[['Sales','Profit']].sum()
```

```python
segment_performence.plot(kind='bar',title="Sales and profit by segment")
plt.show()
```

10. Correlation Analysis.
    Objective: Check for correlations between numerical variables like Sales, Profit, Discount, and Shipping Cost.

```python
import matplotlib.pyplot as plt
```

```python
correlation_matrix=data[['Sales','Profit','Discount','Shipping Cost']].corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix',fontsize=16)
plt.show()
```

11. Boxplot for Profit by Discount Levels
    Use: Visualize the spread and distribution of profit across different discount levels to detect outliers or trends.


```python
plt.figure(figsize=(17,16))
sns.boxenplot(x='Discount',y='Profit',data=data)
plt.title('Profit Distribution by Discount Levels')
plt.show()
```

12. Sales Distribution by Category (Histogram)
    Use: Show the distribution of sales across different product categories.

```python
data['Sales'].hist(by=data['Category'], bins=15, figsize=(10, 8), layout=(2, 2))
plt.suptitle('Sales Distribution by Product Category')
plt.show()
```
